```python
import numpy as np 
import pandas as pd 
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense, Conv2D, Flatten, BatchNormalization, Activation
from keras.utils.np_utils import to_categorical
from keras.callbacks import ReduceLROnPlateau
from keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split
from tensorflow.keras.optimizers import SGD, Adam
```

```python
x_train_full=pd.read_csv('/kaggle/input/digit-recognizer/train.csv', header='infer')
x_test=pd.read_csv('/kaggle/input/digit-recognizer/test.csv', header='infer')
```

```python
x_train=x_train_full.iloc[:,1:]
y_train=x_train_full.iloc[:,0]

x_train
```

```python
x_train=x_train/255.
x_test=x_test/255.
```

```python
# x_train=x_train.values.reshape(-1,28,28,1)
# x_test=x_test.values.reshape(-1,28,28,1)

# x_train.shape
```

```python
y_train_vectors=to_categorical(y_train)
```

```python
x_train, x_val, y_train, y_val= train_test_split(x_train, y_train_vectors, test_size=0.25, random_state=5)
```

```python
print(x_train.shape)
print(x_val.shape)
print(y_train.shape)
print(y_val.shape)
```

```python
model = Sequential()

model.add(Dense(512, input_shape=(784,), activation=None))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Dense(256, input_shape=(512,), activation=None))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Dense(128, input_shape=(256,), activation=None))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Dense(64, input_shape=(128,), activation=None))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Dense(10, input_shape=(64,), activation=None))
model.add(Activation('softmax'))
```

```python
adam = Adam(learning_rate=0.005)

model.compile(loss="categorical_crossentropy", optimizer=adam, metrics=["accuracy"])
```

```python
trained_model = model.fit(x_train,y_train,validation_data=(x_val,y_val), epochs=100, batch_size=50)
```

```python
predictions_final = model.predict(x_test)

print(predictions_final)

predictions_final = np.argmax(predictions_final, axis=1)

print(predictions_final)
```

```python
sub = pd.read_csv('/kaggle/input/digit-recognizer/sample_submission.csv')

sub["Label"] = predictions_final

print(sub)

sub.to_csv('submission_ann.csv',index=False)
```
