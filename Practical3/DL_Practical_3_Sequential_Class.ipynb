```python
from sklearn.preprocessing import LabelBinarizer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.datasets import mnist
from tensorflow.keras import backend as K
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import numpy as np
```

```python
(x_train , y_train),(x_test,y_test) = mnist.load_data()
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_train = x_train.reshape(x_train.shape[0],28*28*1)
x_test = x_test.reshape(x_test.shape[0],28*28*1)
```

```python
x_train.shape
```

```python
x_train
```

```python
x_train[0]
```

```python
x_train = x_train.astype("float32") / 255.0

x_train[0]
```

```python
y_train
```

```python
lb = LabelBinarizer()
y_train = lb.fit_transform(y_train)
y_test = lb.transform(y_test)

y_train
```

```python
y_train.shape
```

```python
model = Sequential()

model.add(Dense(256, input_shape=(784,), activation="relu"))
model.add(Dense(128, input_shape=(256,), activation="relu"))
model.add(Dense(10, input_shape=(128,), activation="softmax"))
```

```python
model.summary()
```

```python
sgd = SGD(learning_rate=0.05)

model.compile(loss="categorical_crossentropy", optimizer=sgd, metrics=["accuracy"])
trained_model = model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=100, batch_size=100)
```

```python
y_test
```

```python
y_test.argmax(axis=1)
```

```python
predictions = model.predict(x_test, batch_size=100)
print(classification_report(y_test.argmax(axis=1), predictions.argmax(axis=1)))
```
